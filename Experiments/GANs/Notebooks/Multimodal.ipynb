```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
means = range(20)
std = 1 + np.random.randn(20)

means = np.array(means)[:, np.newaxis]
std = np.array(std)[:, np.newaxis]
x = np.linspace(-5, 25, 1000)
p = np.mean(np.exp(-(x - means)**2 / (2 * std**2)) / np.sqrt(2 * np.pi * std**2), 0)
plt.plot(x, p, 'r')

eps = np.random.randn(10000)
k = np.random.randint(0, 20, 10000)
x = eps * std[k, 0] + means[k, 0]
plt.hist(x, bins=50, color='white', normed=True)
plt.show()
```
