# Extract Embeddings

## Get GloVe Words

```python
assert np.array([len(datum.split()) for datum in glove]).std() == 0
```

```python
glove_words = [datum.split()[0] for datum in glove]
```

## Fit Tokenizer

```python
vocab_size = 2000
```

```python
tokenizer = Tokenizer(vocab_size)
tokenizer.fit_on_texts([data])
```

## Find Common Words

```python
common_words = [word for word in tokenizer.word_index.keys() if word in glove_words]
```

```python
word_idx = tokenizer.word_index = {common_words[i]: i for i in range(len(common_words))}
```

```python
word_idx_glove = {word: np.where(np.array(glove_words) == word)[0][0]
                  for word in tqdm(word_idx.keys())}
```

## Get Embeddings

```python
glove_embeddings = np.zeros((vocab_size, 50))
```

```python
for i, glove_idx in enumerate(word_idx_glove.values()):
    if i == vocab_size:
        break
    glove_embeddings[i] = np.array([float(n) for n in glove[i].split()[1:]])
```
